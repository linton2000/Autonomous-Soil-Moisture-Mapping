```python
import scipy.io as sp
import pandas as pd
import utils as ut
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import seaborn as sns
```

```python
yancoDf = pd.read_csv('./combined_yanco.csv')
yancoDf
```

```python
def plotSMMap(df):
    smPalette = sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
    fig, ax = plt.subplots()
    sns.scatterplot(data=yancoDf, x='lngDeg_x', y='latDeg_x', hue='Soil Moisture', palette=smPalette)
    plt.title('Soil Moisuture Map of Yanco ELBARA Sampling Area')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    ax.set_xticks(ax.get_xticks()[::2])
    ax.set_yticks(ax.get_yticks()[::2])
```

```python
plotSMMap(yancoDf)
```
